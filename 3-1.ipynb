# 회귀

```python
import pandas as pd
```

```python
df = pd.read_csv('data/Fish.csv')
df.head()
```

```python
cond = df['Species'] == 'Perch'
perch_data = df.loc[cond]
perch_data
```

```python
# 그래프 그리기
import matplotlib.pyplot as plt
# 하나의 특성 사용-> 특성데이터 x, 타깃데이터 y
plt.scatter(perch_data['Length2'], perch_data['Weight'])
```

```python
from sklearn.model_selection import train_test_split 
```

```python
# 사이킷런에 사용하는 훈련세트는 2차원 배열
train_input, test_input, train_target, test_target = train_test_split(perch_data[['Length2']], perch_data[['Weight']])
```

```python
train_input.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor()
knr.fit(train_input, train_target)
```

```python
# 정답을 맞힌 개수의 비율
knr.score(test_input, test_target)
```

```python
knr.predict([[100]]) # 이상하군
```

```python
# 타깃과 예측의 절댓값 오차를 평균하여 반환
from sklearn.metrics import mean_absolute_error
```

```python
# 테스트 세트의 예측 생성
test_prediction = knr.predict(test_input)
print(test_input, test_prediction)
```

```python
# 테스트 세트의 평균 절댓값 오차 게산
mae = mean_absolute_error(test_target, test_prediction)
print(mae)
```

```python
knr.score(train_input, train_target)
```

```python
# 훈련세트는 좋으나, 테스트 세트에서 점수가 나쁘면 모델이 훈련세트에 과대적합 되었다. 
# <-> 반대이거나 두 점수 모두 낮으면 과소적합
# -> 모델을 더 복잡하게 만들면 해결. (훈련세트에 더 잘 맞게 만들면)
```

```python
knr.n_neighbors = 3
knr.fit(train_input, train_target)
knr.score(train_input, train_target)
```

```python
knr.score(test_input, test_target) ## 왜 더 높지 ㅜㅜ
```
